# Watercourse
This feature class contains the watercourse line geometry for the New Brunswick Hydrographic Network. It contains observed streams as well as inferred centrelines representing an approximate centreline path through waterbodies and around islands within the waterbodies.

https://hub.arcgis.com/datasets/NBDNR::nbhn-watercourse-cours-deau-nbhn-0000-01-wc/about

https://www2.gnb.ca/content/gnb/en/departments/erd/open-data/nbhn.html

```
import contextily as cx
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```
gdf = gpd.read_file(
    'https://opendata.arcgis.com/api/v3/datasets/43dc89545eab4bbeaf32c534c39985a7_0/downloads/data?format=shp&spatialRefId=4326&where=1%3D1'
)

gdf = gdf.to_crs('EPSG:2953')
```

```
ax = gdf.plot(
    column='STREAMORDE',
    figsize=(10, 10),
    legend=True,
    cmap='Blues',
    legend_kwds={'location': 'bottom', 'shrink': .8, 'pad': 0.01},
)

cx.add_basemap(ax, crs=gdf.crs.to_string(), source=cx.providers.Esri.NatGeoWorldMap)
ax.set_title('30 Day Map Challenge - Day 2: NB Watercourse (Strahler stream classification value)')
plt.xticks([]), plt.yticks([])
plt.savefig('day02.png', dpi=300, bbox_inches='tight');
```
